```python
from PIL import Image
import matplotlib.pyplot as plt
import numpy as np
from google.colab import files

# Fungsi untuk mengubah ukuran gambar
def resize_image(image, scale=0.5):
    new_width = int(image.width * scale)
    new_height = int(image.height * scale)
    # Menggunakan LANCZOS sebagai pengganti ANTIALIAS
    resized_image = image.resize((new_width, new_height), Image.LANCZOS)
    return resized_image

# Unggah gambar
uploaded = files.upload()

for filename in uploaded.keys():
    # Buka gambar
    original_image = Image.open(filename)

    # Ubah ukuran gambar
    resized_image = resize_image(original_image, scale=0.5)

    # Tampilkan perbandingan
    plt.figure(figsize=(10, 5))

    # Gambar asli
    plt.subplot(1, 2, 1)
    plt.title('Original Image')
    plt.imshow(np.array(original_image))
    plt.axis('off')

    # Gambar yang diubah ukurannya
    plt.subplot(1, 2, 2)
    plt.title('Resized Image')
    plt.imshow(np.array(resized_image))
    plt.axis('off')

    plt.show()

    # Simpan kedua gambar
    original_image.save('original_image.png')
    resized_image.save('resized_image.png')

    # Unduh kedua gambar
    files.download('original_image.png')
    files.download('resized_image.png')
```
